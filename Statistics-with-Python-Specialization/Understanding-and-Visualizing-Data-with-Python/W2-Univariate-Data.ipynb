```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Week 2 - Univariate Data

## Categorical Data

```python
direction = np.random.choice(['up', 'down', 'left', 'right'], 100)
moving = np.random.choice(['y', 'n'], 100)

data = {'ID':[i for i in range(100)],
        'Direction':direction,
        'Moving':moving}

df = pd.DataFrame(data)
df.head()
```

### Fequency Table

```python
df['Moving'].value_counts()
```

```python
df.groupby('Direction').count()['ID']
```

```python
df['Direction'].value_counts(sort=False)
```

```python
(df['Direction'].value_counts()/len(df['Direction']))*100
```

```python
pd.crosstab(index=df['Moving'], columns=df['Direction'])
```

### Bar Chart

```python
df.groupby('Direction').count()['ID'].plot(kind='bar')
plt.show()
```

```python
plt.bar(np.arange(4), df['Direction'].value_counts())
plt.xticks(np.arange(4), df['Direction'].value_counts().index)
plt.show()
```

### Histograms

- Shape

    - Unimodal - only one peak
    - Bomodal
- Center
    - Mean or median
- Spread
    - Range (max - min), Interquartile Range (IQR), standard deviation, variance
- Outliers

```python
np.random.rand(20)
```

```python
length = np.random.randint(40, size=100)
weight = length * np.random.random()
data = {'ID':[i for i in range(100)],
        'Weight':weight,
        'Length':length}

df = pd.DataFrame(data)
df.head()
```

```python
df['Length'].hist(bins=3)
plt.show()
```

#### Example summary
To summarize that in one sentence, we might get something like the distribution of salaries in San Francisco is bimodal and skewed to the right.  Centered at about \$80,000 with most of the data between \$40,000 and \$120,000.  A range of roughly \$600,000, and outliers are present on the higher end.

## Standard Score (Emperical Rule)
- Bell-shaped (normal) distributions
- Standard Deviation - roughly the average distance of the values from the mean
- 68% of values within +- one SD 
- 95% of values within +- two SD
- 99.7% of values within +- three SD 
- z-score = (x - mean) / SD
